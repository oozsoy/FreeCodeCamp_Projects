```python
import copy
import random


class Hat:
    
    def __init__(self, **kwargs): # kwargs to pass in keyworded arguments as yellow=2 with as many 
                                  # arguments as you want
                                
        self.contents = []
  
        for key, val in kwargs.items():      # seperate the input to keys and values 
          self.contents += val * [key]    # multiply each key with the value and append to contents
  
    def draw(self, num_b):
        
        num_b = min(num_b, len(self.contents)) # number of balls drawn should be <= full content in                                                # the hat
      
        balls_drawn = []

          
        for i in range(num_b): # pop random elements from the total content and append to a list
                             # this gives the drawn balls for the number specfified
        
            balls_drawn.append(self.contents.pop(random.randrange(len(self.contents))))
    
        return balls_drawn          
    
def experiment(hat, expected_balls, num_balls_drawn, num_experiments): # expected_balls is a dictionary
    
    M = 0 # number of occurance when expected_balls \equiv balls drawn
  
    for i in range(num_experiments): # loop through the number of experiments 
        
        hat_copy = copy.deepcopy(hat)
        
        balls_drawn = hat_copy.draw(num_balls_drawn)

        balls_req = 0 # number of different requested colors/types
      
        for key, val in expected_balls.items(): 
            
            
            
            if balls_drawn.count(key) >= val:  # If the number of balls
                                               # with a given color is more than or equal to values in expected_balls
                                               # add +1 to the control variable balls_req which gives the total 
                                               # number of colors/type two lists / dictionaries match
                balls_req += 1
                
        if balls_req == len(expected_balls):   # since each type of color within expected balls
                                               # should be also included in the drawn list, balls_req = len(expected_balls)
            
            M += 1
            
        else:        
            
            0
        
    probability = M / num_experiments
    return probability
      
       
```
